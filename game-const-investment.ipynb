```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P = 0.5
ROI = 0.1 # e.g. invest 40, get 4
INVEST_RATIO = 0.02
```

```python
def play_game(amount_start = 100, nb_rounds = 1, invest_ratio = INVEST_RATIO):
    
    amount_current = amount_start
    nb_rounds_till_end = 0
    amount_history = [amount_current]
    amount_invest = amount_current * (invest_ratio)
    
    for i in range(nb_rounds):
        
        amount_return = amount_invest * (ROI + 1)
        
        if amount_current <= 1:
            
            break
            
        nb_rounds_till_end += 1
        
        if P > np.random.uniform(size = 1):
            
            amount_current = amount_current + amount_return
            
        else:
            
            amount_current = amount_current - amount_invest
        
        amount_history.append(amount_current)
        
    return amount_current, nb_rounds_till_end, amount_history
    
```

```python
def play_games(nb_games = 100, amount_start = 100, invest_ratio = INVEST_RATIO, nb_rounds_per_game = 1000):
    
    roi_absolute = 0
    nb_rounds = 0
    history = []
    
    print("No investment amount available anymore in round(s):")
    for i in range(nb_games):
        
        r, n, h = play_game(amount_start, nb_rounds_per_game, invest_ratio)
        roi_absolute += r
        nb_rounds += n
        
        if n < nb_rounds_per_game:
            print(i, end = " ")
        
        history.append(h)
        
    avg_roi_absolute = roi_absolute / nb_games
    avg_rounds = nb_rounds / nb_games
    
    return avg_roi_absolute, avg_rounds, history
```

```python
avg_roi_abs, avg_rounds, hist = play_games(invest_ratio = 0.2, nb_rounds_per_game = 1000)
```

```python
print(avg_roi_abs, avg_rounds)
```

```python
for i in range(5):
    plt.plot(hist[i])
    plt.show()
```
